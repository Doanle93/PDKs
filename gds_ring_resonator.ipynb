```python
import gdsfactory as gf
import gplugins.gmeep as gm
import gplugins


c = gf.Component("ring_resonator")
ring = c << gf.components.ring(radius=2.5, width=0.25, angle_resolution=2.5, layer='WG', angle=360)
straight = c << gf.components.straight(width=0.25,length=10)
straight.move([-5, 2.5 + 0.13 + 0.25])
c.add_port("o1", port=straight.ports["o1"])
c.add_port("o2", port=straight.ports["o2"])

# c = gf.components.mmi1x2_with_sbend(with_sbend=True, cross_section="xs_sc")

c.plot()

core_material = gplugins.get_effective_indices(
    core_material=1.5,
    clad_materialding=1.0,
    nsubstrate=1.0,
    thickness=0.22,
    wavelength=0.55,
    polarization="te",
)[0]
```

```python
sp = gm.write_sparameters_meep(c, ymargin_top=1, ymargin_bot=1,run=False)
```

```python
sp = gm.write_sparameters_meep(
    c, 
    resolution= 20, 
    is_3d=False,
    wavelength_start=0.5,
    wavelength_stop=0.6,
    material_name_to_meep = dict(si=core_material),
    filepath="data/ring_resonator.npz"
)
```

```python
gplugins.plot.plot_sparameters(sp)
```

```python
gplugins.plot.plot_sparameters(sp, keys=("o1@0,o2@0",))
```

```python
import gdsfactory as gf
import gplugins.gmeep as gm
import gplugins

c = gf.Component("ring_resonator")
ring = c << gf.components.ring(radius=2.5, width=0.25, angle_resolution=2.5, layer='WG', angle=360)
straight1 = c << gf.components.straight(width=0.25,length=10)
straight1.move([-5, 2.5 + 0.13 + 0.25])

straight2 = c << gf.components.straight(width=0.25, length=10)
straight2.move([-5, -2.5 - 0.13 - 0.25])

c.add_port("o1", port=straight1.ports["o1"])
c.add_port("o2", port=straight1.ports["o2"])
c.add_port("o3", port=straight2.ports["o1"])
c.add_port("o4", port=straight2.ports["o2"])


# c = gf.components.mmi1x2_with_sbend(with_sbend=True, cross_section="xs_sc")

c.plot()

core_material = gplugins.get_effective_indices(
    core_material=1.5,
    clad_materialding=1.0,
    nsubstrate=1.0,
    thickness=0.22,
    wavelength=0.545,
    polarization="te",
)[0]
```

```python
sp = gm.write_sparameters_meep(c, ymargin_top=1, ymargin_bot=1,run=False)
```

```python
sp = gm.write_sparameters_meep(
    c, 
    resolution= 20, 
    is_3d=False,
    wavelength_start=0.5,
    wavelength_stop=0.6,
    material_name_to_meep = dict(si=core_material),
    filepath="data/ring_resonator2.npz"
)
```

```python
gplugins.plot.plot_sparameters(sp)
```

```python
gplugins.plot.plot_sparameters(sp, keys=("o2@0,o1@0",))
```
